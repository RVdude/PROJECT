## Sector-Wide Stock Market Prediction Using Time Series Analysis on Brazilian Equities

### Data Integration and Preprocessing

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Load the dataset
dataset = pd.read_csv('brazallian_stocks_prediction.csv')
```

```python
dataset.info()
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
# Convert Date to datetime
dataset['Date'] = pd.to_datetime(dataset['Date'])
```

```python
# Pivot the dataset to have stock names as columns and dates as rows
pivot_df = dataset.pivot(index='Date', columns='Stock Name', values='Close')
```

```python
# Fill missing values (interpolation or forward fill)
pivot_df = pivot_df.interpolate(method='time').fillna(method='bfill').fillna(method='ffill')
```

```python
# Normalize the data
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(pivot_df)
```

```python
# Convert scaled data back to DataFrame
scaled_df = pd.DataFrame(scaled_data, index=pivot_df.index, columns=pivot_df.columns)
```

```python
# Save the preprocessed data for further use
scaled_df.to_csv('preprocessed_brazillian_stocks_predictions_data.csv')
```

### Model Building

#### VAR Model 

```python
from statsmodels.tsa.api import VAR

# Fit the VAR model
model_var = VAR(scaled_df)
results_var = model_var.fit(maxlags=15, ic='aic')

# Print the summary of the model
print(results_var.summary())

# Forecasting
forecast_var = results_var.forecast(scaled_df.values[-results_var.k_ar:], steps=10)
forecast_var_df = pd.DataFrame(forecast_var, index=pd.date_range(start=scaled_df.index[-1], periods=10, freq='B'), columns=scaled_df.columns)
```

```python
# Plot the forecasted results
plt.figure(figsize=(14, 7))
for column in forecast_var_df.columns:
    plt.plot(forecast_var_df.index, forecast_var_df[column], label=column)
plt.title('VAR Model Forecast')
plt.xlabel('Date')
plt.ylabel('Forecasted Prices')
plt.legend()
plt.show()
```
